```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("residential_housing_updated.csv")
```

```python
#3
missing_per_column = df.isna().sum()
total_missing = missing_per_column.sum()

print(missing_per_column)
print("Total missing values:", total_missing)
```

```python
#4a
# Select only numeric columns for median computation
numeric_columns = df.select_dtypes(include=['number']).columns

# Compute the median for numeric columns
medians = df[numeric_columns].median()
print("Median:" , medians)

# Compute the mean for numeric columns
mean = df[numeric_columns].mean()
print("Mean:" , mean)

# Compute the mode for numeric columns
mode = df[numeric_columns].mode()
print("Mode:" , mode)

standard_deviation = df[numeric_columns].std()
print("Standard Deviation:",standard_deviation)
```

```python
#4b

# Fill missing values in numeric columns with their respective medians
df_filled_median = df.copy()
df_filled_median[numeric_columns] = df_filled_median[numeric_columns].fillna(medians)

missing_per_column = df_filled_median.isna().sum()
total_missing = missing_per_column.sum()

print(missing_per_column)
print("Total missing values:", total_missing)
```

```python
#4c
# Select only numeric columns for median computation
numeric_columns = df_filled_median.select_dtypes(include=['number']).columns

# Compute the median for numeric columns
medians = df_filled_median[numeric_columns].median()
print("Median:" , medians)

# Compute the mean for numeric columns
mean = df_filled_median[numeric_columns].mean()
print("Mean:" , mean)

# Compute the mode for numeric columns
mode = df_filled_median[numeric_columns].mode()
print("Mode:" , mode)

standard_deviation = df_filled_median[numeric_columns].std()
print("Standard Deviation:",standard_deviation)
```

```python
# Replace missing values by propagating previous non-missing values
df_filled_propagated = df.fillna(method='ffill')

# Display the first few rows of the filled DataFrame
print("DataFrame after forward filling missing values:")
print(df_filled_propagated.head())

# Check if there are still missing values
missing_values_count = df_filled_propagated.isnull().sum().sum()
print(f"Total number of missing values after forward filling: {missing_values_count}")
```
